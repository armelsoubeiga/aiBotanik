```python
import os

def create_file(filepath):
    """Cr√©e un fichier vide"""
    try:
        with open(filepath, 'w', encoding='utf-8') as f:
            pass  # Fichier vide
        print(f"‚úì Fichier cr√©√©: {filepath}")
    except Exception as e:
        print(f"‚úó Erreur lors de la cr√©ation de {filepath}: {e}")

def create_directory(dirpath):
    """Cr√©e un r√©pertoire s'il n'existe pas"""
    try:
        os.makedirs(dirpath, exist_ok=True)
        print(f"‚úì Dossier cr√©√©: {dirpath}")
    except Exception as e:
        print(f"‚úó Erreur lors de la cr√©ation du dossier {dirpath}: {e}")

def create_botanikai_structure():
    """Cr√©e la structure compl√®te du projet botanikai"""
    
    print("üå± Cr√©ation de la structure du projet aiBotanik...")
    print("=" * 50)
    
    # Structure des dossiers
    directories = [
        "botanikai",
        "botanikai/backend",
        "botanikai/backend/.github",
        "botanikai/backend/.github/workflows",
        "botanikai/frontend",
        "botanikai/frontend/src",
        "botanikai/frontend/src/components"
    ]
    
    # Cr√©ation des dossiers
    print("\nüìÅ Cr√©ation des dossiers...")
    print("-" * 30)
    for directory in directories:
        create_directory(directory)
    
    # Liste des fichiers √† cr√©er (vides)
    files = [
        # Fichiers racine
        "botanikai/logo.svg",
        "botanikai/README.md",
        
        # Backend
        "botanikai/backend/app.py",
        "botanikai/backend/langchain_chains.py",
        "botanikai/backend/requirements.txt",
        "botanikai/backend/.github/workflows/hf_deploy.yml",
        
        # Frontend
        "botanikai/frontend/package.json",
        "botanikai/frontend/vercel.json",
        "botanikai/frontend/src/App.jsx",
        "botanikai/frontend/src/api.js",
        "botanikai/frontend/src/index.js",
        "botanikai/frontend/src/components/SearchForm.jsx",
        "botanikai/frontend/src/components/Recommendation.jsx"
    ]
    
    # Cr√©ation des fichiers vides
    print("\nüìÑ Cr√©ation des fichiers...")
    print("-" * 30)
    for filepath in files:
        create_file(filepath)
    
    print("\n" + "=" * 50)
    print("‚úÖ Structure du projet aiBotanik cr√©√©e avec succ√®s!")
    print("üìä R√©sum√©:")
    print(f"   - {len(directories)} dossiers cr√©√©s")
    print(f"   - {len(files)} fichiers cr√©√©s")
    print("\nüöÄ Votre projet est pr√™t √† √™tre d√©velopp√©!")
    print("üìÇ Dossier racine: botanikai/")

if __name__ == "__main__":
    # V√©rifier si on est dans le bon r√©pertoire
    current_dir = os.path.basename(os.getcwd())
    if current_dir != "aiBotanik":
        print("‚ö†Ô∏è  Attention: Ce script doit √™tre ex√©cut√© depuis le dossier 'aiBotanik'")
        print(f"   R√©pertoire actuel: {os.getcwd()}")
        choice = input("Voulez-vous continuer quand m√™me? (o/n): ")
        if choice.lower() != 'o':
            print("‚ùå Cr√©ation annul√©e")
            exit()
    
    create_botanikai_structure()
```
